```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
def mysql_connection(host, user, passwd, database=None):
    engine = create_engine(f'mysql+pymysql://{user}:{passwd}@{host}/{database}')
    return engine.connect()

mysql = mysql_connection('127.0.0.1', 'root', '123456', 'pnad_covid')
```

```python
df = pd.read_csv('PNAD_COVID_052020.csv', sep=';')
df.head()
```

```python
df.to_sql('dados', mysql, if_exists='replace', index=False)
```
